# **<b>Exercise 1:</b> Input a character from user and use the <b>if</b> statement to print corresponding types of the character (small letter, capital letter, digit, punctuation).**

```python
# code here
import string

a = input("Enter an input: ")

if a.isalpha():
    if a.isupper():
        print(f"The input >>> {a} <<< is in uppercase.")
    else:
        print(f"The input >>> {a} <<< is in lowercase.")

elif a.isdigit():
    print(f"The input >>> {a} <<< is a digit.")

elif a in string.punctuation:
    print(f"The input >>> {a} <<< is a punctuation character.")

else:
    print("Please enter a valid input.")
```

# Exercise 4: Check Palindrome Number

```python
#list = input("Enter a Palindrome: ").split()
#print(list)

def is_palindrome(text):
    # Step 1 & 2: Convert to lowercase and keep only alphanumeric characters
    cleaned_text = "".join(char.lower() for char in text if char.isalnum())
    
    # Step 3: Compare the cleaned string with its reverse
    return cleaned_text == cleaned_text[::-1]

# Testing the function
user_input = input("Enter a word or phrase: ")

if is_palindrome(user_input):
    print("Yes, it's a palindrome!")
else:
    print("No, it's not a palindrome.")


             
```

# for statement
<p style="text-align: justify">The for statement in Python differs a bit from what you may be used to in C or C++. Rather than always giving the user the ability to define both the iteration step and halting condition, Python’s for statement iterates over the items of any sequence (a list or a string), in the order that they appear in the sequence.</p>

```python
# Measure some strings:
words = ['cat', 'window', 'dhaka', 'jack']
for w in words:
    print(w, len(w))
```

```python
i = 1
while i < 6:
    print(i)
    i += 1
```

```python
numbers = [1, 56, 78, 8,7]

for n in numbers:
    print(n+10)
```

### The range() Function
If you do need to iterate over a sequence of numbers, the built-in function range() comes in handy. It generates arithmetic progressions:

```python
# repeated for 0 to 4
# for(i=0;i<5;i++)
for i in range(5):
    print(i)
```

<p style="text-align: justify">The given end point is never part of the generated sequence; range(10) generates 10 values, the legal indices for items of a sequence of length 10. It is possible to let the range start at another number, or to specify a different increment (even negative; sometimes this is called the ‘step’):</p>

```python
list(range(5, 10))
```

```python
list(range(0, 10, 3))
```

```python
list(range(10, 2, -1))
```

```python
list(range(-10, -100, -30))
```

```python
for i in range(len(numbers)):
    print(i, numbers[i])
```

```python
for i in range(2,100,5):
    print(i,end=',')
```

# Functions
<p style="text-align: justify">The keyword def introduces a function definition. It must be followed by the function name and the parenthesized list of formal parameters. The statements that form the body of the function start at the next line, and must be indented.</p>

```python
def operate(a, b):
  print(str(a) + ' + ' + str(b) + ' = ' + str(a+b))
  print(str(a) + ' - ' + str(b) + ' = ' + str(a-b))
  print(str(a) + ' x ' + str(b) + ' = ' + str(a*b))
  print(str(a) + ' / ' + str(b) + ' = ' + str(a/b))
```

```python
operate(91,5)
```

```python
operate(5,2)
```

```python
def fib(n):    # write Fibonacci series up to n
    """Print a Fibonacci series up to n."""
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
        #print()
```

```python
# Now call the function we just defined:
fib(50)
```

```python
# define method
def addnumbers(a, b): # add numbers and return result
    # logic
    c = a + b
    return c
```

```python
y = addnumbers('8ppp', '9') # use or call method
print(y)
```

```python
addnumbers([33,77], [66])
```

```python
# add list values
def addValues(num):
    sum = 0
    for n in num:
        sum = sum + n
    return sum
```

```python
addValues([])
```

```python
addValues((9,7,5))
```

```python
v = [5, 4,12, 7]
addValues(v)
```

```python
# method/function
def multiTable(n, r=10):
    for i in range(1,r+1):
        print("{}x{}={}".format(n,i,n*i))
```

```python
multiTable(8)
```

```python
multiTable(15)
```

```python
v = [5, 4,12, 7, 5, 4,12, 7, 5, 4,12, 7]
addValues(v)
```

<b>Exercise 4:</b> Define a <b>method/function</b> which will take two lists of numbers and returns a list (sum of two lists)
<p>Example:<br>
    Input: [3,4,5,1] and [6,7,2,8]<br>
    output: [9,11,7,9]</p>


```python
# code here
input1 = input("Enter list of numbers saperated by spaces for list1: ")
input2 = input("Enter list of numbers saperated by spaces for list2: ")

list1 = []
for i in input1.split():
    x = int(i)
    list1.append(x)

list2 = []
for i in input2.split():
    x = int(i)
    list2.append(x)
    
list3 = []

for i in range(len(list1)):
    t = list1[i] + list2[i]
    list3.append(t)
for i in list3:
    print(i, end = ", ")
```

<b> Exercise 5:</b> Write a <b>method</b> which will take the <b>full name</b> as input and return both the first name and last name.
<p>Example:<br>
    Input: Arafat Rahman Sunny<br>
    Output: First name: Arafat, Last name: Sunny</p>

```python
# code here
inputn = input("Enter your name: ")
namelist = []
for i in inputn.split():
    x = i
    namelist.append(x)
print("First Name: " + namelist[0])
print("Last Name: " + namelist[-1])
    
```
