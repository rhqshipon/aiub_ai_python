```python
graph = {
    'University': {'A': 5, 'B': 7},
    'A': {'C': 3, 'D': 6},
    'B': {'E': 2, 'F': 5},
    'C': {'Airport': 4},
    'D': {'Airport': 2},
    'E': {'Airport': 3},
    'F': {'Airport': 1},
    'Airport': {}
}

h = {
    'University': 12, 
    'A': 8, 
    'B': 7, 
    'C': 5, 
    'D': 4, 
    'E': 3, 
    'F': 2, 
    'Airport': 0
}
```

```python
def uniform_cost_search(start, goal):
    print("--- UNIFORM COST SEARCH (UCS) ---")
    queue = [(0, start, [start])]
    visited = [] 
    iteration = 1
    
    while queue:
        queue.sort()
        
        pq_display = [(cost, node) for cost, node, _ in queue]
        print(f"Iteration {iteration} Priority Queue: {pq_display}")
        
        cost, current, path = queue.pop(0)
        
        if current in visited:
            iteration += 1
            continue
            
        visited.append(current)
        
        if current == goal:
            print(f"\nOrder of node expansion: {' -> '.join(visited)}")
            print(f"Least-cost path: {' -> '.join(path)}")
            print(f"Total path cost: {cost}\n")
            return
            
        for neighbor, edge_cost in graph[current].items():
            if neighbor not in visited:
                queue.append((cost + edge_cost, neighbor, path + [neighbor]))
        iteration += 1
```

```python
def greedy_best_first_search(start, goal):
    print("--- GREEDY BEST-FIRST SEARCH (GBFS) ---")
    current = start
    path = [start]
    total_cost = 0
    expanded = []
    
    while current != goal:
        expanded.append(current)
        neighbors = graph[current]
        
        next_node = min(neighbors, key=lambda n: h[n])
        
        total_cost += neighbors[next_node]
        current = next_node
        path.append(current)
        
    expanded.append(goal)
    print(f"Order of node expansion: {' -> '.join(expanded)}")
    print(f"Final path: {' -> '.join(path)}")
    print(f"Total path cost: {total_cost}\n")
```

```python
def a_star_search(start, goal):
    print("--- A* SEARCH ---")
    queue = [(h[start], start, 0, [start])]
    visited = []
    iteration = 1
    
    while queue:
        queue.sort()
        
        f_cost, current, g_cost, path = queue.pop(0)
        
        if current in visited:
            continue
            
        visited.append(current)
        
        
        print(f"Iteration {iteration} - Expanded: {current} | g(n) = {g_cost} | h(n) = {h[current]} | f(n) = {f_cost}")
        
        if current == goal:
            print(f"\nOptimal path: {' -> '.join(path)}")
            print(f"Total path cost: {g_cost}\n")
            return
            
        for neighbor, edge_cost in graph[current].items():
            if neighbor not in visited:
                new_g = g_cost + edge_cost
                new_f = new_g + h[neighbor]
                queue.append((new_f, neighbor, new_g, path + [neighbor]))
        iteration += 1
```

```python
start_node = 'University'
goal_node = 'Airport'

uniform_cost_search(start_node, goal_node)
greedy_best_first_search(start_node, goal_node)
a_star_search(start_node, goal_node)
```
